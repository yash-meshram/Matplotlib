# Histrogram

```python
import numpy as np
```

```python
population_age = np.random.randint(12,120,89)
bins = np.linspace(0,130,14)
print(population_age)
print(bins)
print(np.min(population_age))
print(np.max(population_age))

plt.hist(population_age,bins,histtype='bar',rwidth=0.8)

plt.title('Histrogram')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

# Area graph / Stack plot

```python
days = [1,2,3,4,5]

s = sleeping = [7,8,6,11,7]
e = eating = [2,3,4,3,2]
w = working = [7,8,7,2,2]
p = playing  = [8,5,7,8,13]

# plt.plot([],[],color = 'm',label='Sleeping',linewidth=5)
# plt.plot([],[],color = 'c',label='Eating',linewidth=5)
# plt.plot([],[],color = 'r',label='Working',linewidth=5)
# plt.plot([],[],color = 'k',label='Playing',linewidth=5)

plt.stackplot(days,s,e,w,p,colors=['m','c','r','k'],labels=['Sleeping',
                                                            'Eating',
                                                            'Working',
                                                            'Playing']) #colors not color

plt.legend()

plt.show()
```

# Pie chart

```python
slices = [7,2,2,13]
activities = ['snacks','laundry','internet','food']
colors = ['c','m','r','b']

plt.pie(slices,
       colors=colors,
       labels=activities,
       startangle=90,
       shadow=True,
       explode=(0,0.15,0,0),
       autopct='%1.5f%%') #% kese display hoo .n means n digit after point

plt.title('Pie Chart')
plt.show()
```

# Working with multiple plots

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.02)

plt.subplot(211) #plt.subplot(vertically we have 2 plots  horizontally we have1 plot  1st graph)
plt.plot(t1,f(t1),'bo',t2,f(t2),color='b')
plt.show()

plt.subplot(121) #plt.subplot(vertically we have 2 plots  horizontally we have1 plot  1st graph)
plt.plot(t1,f(t1),'bo',t2,f(t2),color='b')
plt.show()

plt.subplot(221) #plt.subplot(vertically we have 2 plots  horizontally we have1 plot  1st graph)
plt.plot(t1,f(t1),'bo',t2,f(t2),color='b')
plt.show()

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),color='b')
plt.show()

plt.subplot(122)
plt.plot(t2,np.cos(2*np.pi*t2),color='b')
plt.show()

plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2),color='b')
plt.show()
```

```python
#importing libraries

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


# setting up figure size and DPI for screen 

%matplotlib inline
plt.rcParams['figure.figsize'] = (6,4) #size, how much horizontal and vertical strech in image only and not in value (hor.,ver.)
plt.rcParams['figure.dpi'] = 150  #DPI= Dot Per Inches

#  Adding a 3D Scatterplot

x = np.random.normal(size=500)
y = np.random.normal(size=500)
z = np.random.normal(size=500)

a = plt.subplot(projection='3d')

# a.scatter(x,y,z)
# plt.show()

a.scatter(x,y,z,c = np.linalg.norm([x,y,z],axis=0))
plt.show()
```
